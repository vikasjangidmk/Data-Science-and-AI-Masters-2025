```python
# Seaborn for plotting and styling
import seaborn as sns
df = sns.load_dataset('iris')
print(df.head())
```

```python
len(df)
```

```python
df.species.value_counts()
```

```python
df.species.value_counts().plot(kind='bar')
```

```python
sns.countplot(data=df, x='species', order=df.species.value_counts().index)
```

```python
sns.color_palette()
```

```python
sns.countplot(data=df, x='species', order=df.species.value_counts().index, palette='flare')
```

```python
sns.kdeplot(df['sepal_length'])
```

```python
sns.kdeplot(df['sepal_length'], shade = True, color='green')
```

```python
# KDE: kernel density estimation
for col in ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']:
    sns.kdeplot(df[col], shade=True)
```

#### Histograms + KDE = Distplot

```python
sns.displot(df['sepal_length'])
```

```python
sns.histplot(df['sepal_length'])
```

```python
sns.pairplot(df, hue='species', size=2.5);
```
